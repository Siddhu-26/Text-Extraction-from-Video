```python
from PIL import Image
import imagehash
import os
import numpy as np

class DuplicateRemover:
    def __init__(self,dirname,hash_size = 8):
        self.dirname = dirname
        self.hash_size = hash_size
        
    def find_duplicates(self):
        """
        Find and Delete Duplicates
        """
        
        fnames = os.listdir(self.dirname)
        hashes = {}
        duplicates = []
        print("Finding Duplicates Now!\n")
        for image in fnames:
            with Image.open(os.path.join(self.dirname,image)) as img:
                temp_hash = imagehash.average_hash(img, self.hash_size)
                if temp_hash in hashes:
                    duplicates.append(image)
                else:
                    hashes[temp_hash] = image
                   
        c = 0
        if len(duplicates) != 0:
            for duplicate in duplicates:                    
                os.remove(os.path.join(self.dirname,duplicate))
                c += 1
            
            print("{} Images Deleted Succesfully!".format(c))
        else:
            print("No Duplicates Found :(")

                    

# Remove Duplicates
dr = DuplicateRemover("E:\Programs\Python\Text_Image\data")
dr.find_duplicates()
```

```python
# Extracting the text from the images
os.chdir("E:\Programs\Python\Text_Image\data")
final = []
rep = []

path_to_extract = []
for (root,dirs,file) in os.walk(os.getcwd(), topdown = False):
    path_to_extract = (file)

path_to_extract = natsorted(path_to_extract)

print(f"Number of images: {len(path_to_extract)}")

for img in path_to_extract:
    image = Image.open(img)
    image = image.convert('L') # Converting RGB to Greyscale for better recognition

    img_text = pytesseract.image_to_string(image, lang= 'eng') 

    img_text = img_text.strip()
    
    if img_text in final:
        if img_text not in rep:
            rep.append(img_text)
    else:
        final.append(img_text)

         
#final = list(set(final))

#print(final)
```
